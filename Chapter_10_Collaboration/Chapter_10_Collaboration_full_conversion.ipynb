```python
# Docstring and example
# Demonstration: accessing docstrings and using pydoc help
def find_anagrams(word, dictionary):
    """Find all anagrams for a word.

    Args:
        word: target word (str)
        dictionary: a container of known words (iterable of str)
    Returns:
        list of anagrams found (list[str])
    """
    result = []
    for w in dictionary:
        if sorted(w) == sorted(word):
            result.append(w)
    return result

# Show the docstring programmatically
print(find_anagrams.__doc__)

# Small demonstration
words = ['cat', 'act', 'tac', 'dog', 'odg']
print(find_anagrams('cat', words))
```

```python
# Root exception hierarchy example
# Root exception pattern: define a root exception and specific subclasses
class Error(Exception):
    """Base-class for all exceptions raised by this module."""
    pass

class InvalidDensityError(Error):
    """There was a problem with a provided density value."""
    pass

class InvalidVolumeError(Error):
    """There was a problem with the provided volume value."""
    pass

def determine_weight(volume, density):
    if density < 0:
        raise InvalidDensityError('Density must be positive')
    if volume < 0:
        raise InvalidVolumeError('Volume must be positive')
    if volume == 0:
        density / volume  # deliberate ZeroDivisionError

try:
    determine_weight(0, 1)
except Error as e:
    print('Caught module Error:', type(e).__name__, e)
except Exception as e:
    print('Caught unexpected Exception:', type(e).__name__, e)
```

```python
# warnings usage and test capture
# warnings example: issuing deprecation warnings and catching them in tests
import warnings

def require(name, value, default):
    if value is not None:
        return value
    warnings.warn(f'{name} will be required soon, update your code', DeprecationWarning, stacklevel=3)
    return default

# Example usage (this will issue a warning)
found = require('speed_units', None, 'mph')

# Capture warnings programmatically (like in tests)
with warnings.catch_warnings(record=True) as found_warnings:
    warnings.simplefilter('always')  # ensure warnings are recorded
    _ = require('my_arg', None, 'fake units')
    assert len(found_warnings) == 1
    print('Recorded warning message:', str(found_warnings[0].message))
```

```python
# typing and generics example
# typing + mypy-friendly examples (static checkers run outside of runtime)
from typing import Optional, List, Callable, TypeVar

Value = TypeVar('Value')
Func = Callable[[Value, Value], Value]

def combine(func: Func[Value], values: List[Value]) -> Value:
    assert len(values) > 0
    result = values[0]
    for next_value in values[1:]:
        result = func(result, next_value)
    return result

def add(x: int, y: int) -> int:
    return x + y

inputs = [1, 2, 3, 4]  # keep homogeneous types to satisfy static checkers
print('combine(add, inputs) ->', combine(add, inputs))
```
